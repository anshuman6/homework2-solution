```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sensitivity1 = .50 # P(T+|D+)
```

```python
sensitivity2 = 0.7 # P(T+|D+)
```

```python
specificity = .995 # P(T-|D-)
```

```python
pdplus = np.arange(0,1,0.01)
```

```python
pdnegative = 1 - pdplus
```

```python
# 1:P(D+|T+) = P(T+|D+)*P(D+) Divided By P(T+)
# 2:P(T+) = P(T+, D+) + P(T+, D-)
# 3:P(T+) = P(T+|D+)P(D+) + P(T+|D-)P(D-)
# Using above
# P(D+|T+) = P(T+|D+)*P(D+) Divided By P(T+|D+)P(D+) + P(T+|D-)P(D-)
# P(T+|D-) = 1- P(T-|D-) (This is because when Disease is negative, there will be only two cases -> T+ or T-)
```

```python
# Plot P(D+|T+)
```

```python
# at 50% sensitivity
pdplustplus1 = (sensitivity1 * pdplus) / ((sensitivity1 * pdplus) + ((1-specificity)*pdnegative))
# at 70% sensitivity
pdplustplus2 = (sensitivity2 * pdplus) / ((sensitivity2* pdplus) + ((1-specificity)*pdnegative))
```

```python
plt.figure(figsize = (20,5))
plt.subplot(1,2,1)
plt.plot(pdplus,pdplustplus1, 'b', pdplus,pdplustplus2,'r')
plt.xlabel("P(D+) Prevalence")
plt.ylabel("P(D+|T+)")
plt.title("Blue - 50% sensitivity, Red - 70% Sensitivity")
plt.show()
```

### From above we can see that, the higher the sensitivity, the more probability of having disease if test is positive
### This is specially true at lower prevalances
### So ideally sensitivity of test should be as high as possible, ideally 100% or 1.

```python
# Plot P(D+|T-)
```

```python
# at 50% sensitivity
pdplustminus1 = ((1-sensitivity1)*pdplus)/(((1-sensitivity1)*pdplus)+((1-pdplus)*specificity))
# at 70% sensitivity
pdplustminus2 = ((1-sensitivity2)*pdplus)/(((1-sensitivity2)*pdplus)+((1-pdplus)*specificity))
```

```python
plt.figure(figsize = (20,5))
plt.subplot(1,2,1)
plt.plot(pdplus,pdplustminus1, 'b', pdplus,pdplustminus2, 'r')
plt.xlabel("P(D+) Prevalence")
plt.ylabel("P(D+|T-)")
plt.title("Blue - 50% sensitivity, Red - 70% Sensitivity")
plt.show()
```

### At higher prevalence, chance of having disease .. even when test is negative is HIGH
### However, As the sensitivity of test increases, 
### chance of having disease when test is negative is lower - as compared with test of low sensitivity
